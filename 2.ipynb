```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
imgOriginal = cv.imread("/home/shahin/Desktop/M1/Sajjad project/Assignment-27/image/apple.jpg")
imgGray = cv.cvtColor(imgOriginal, cv.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap="gray")
```

```python
_ ,img = cv.threshold(imgOriginal, 220, 240, cv.THRESH_BINARY)
plt.imshow(img, cmap="gray")
```

```python
row, col = imgGray.shape
sign_j = []

for j in range(col):
    flag = False
    list_new = []
    for i in range(row):
        if imgGray[i, j] != 0 :
            if  flag==False:
                flag = True
                list_new.append([i, j])

        elif imgGray[i, j]==0 :
            if flag==True:

                flag = False
                list_new.append([i, j])
                
    if len(list_new) != 0:
        sign_j.append(list_new[0])
        sign_j.append(list_new[-1])
for m in sign_j:
    cv.circle(imgGray, (m[1], m[0]), 3, (110,0,0), 4)
```

```python
plt.imshow(imgGray)
cv.imwrite("result.jpg" , imgGray)
```
