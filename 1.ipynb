```python
imgOriginal = cv.imread("/home/shahin/Desktop/M1/Sajjad project/Assignment-27/image/1.jpg", cv.IMREAD_GRAYSCALE)
_ ,img = cv.threshold(imgOriginal, 140, 255, cv.THRESH_BINARY)
```

```python
contours, _ = cv.findContours(img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
```

```python
imgRes = img.copy()
```

```python
for c in contours:
    if len(c) > 200:
        x, y, w, h = cv.boundingRect(c)
        cv.rectangle(imgOriginal, (x, y), (x+w, y+h),(127, 127, 127), 2)

plt.imshow(imgOriginal, cmap="gray")
```
